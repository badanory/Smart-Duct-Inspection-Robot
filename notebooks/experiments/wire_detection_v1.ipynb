```python
from ultralytics import YOLO

from IPython.display import display, Image # Changed 'Ipython' to 'IPython'
```

```python
# 2) 환경변수에 Roboflow Private API Key 설정
import os
os.environ["ROBOFLOW_API_KEY"] = "1R0h7dpWnUesIXZTWQnf"  # 복사해 둔 Private Key

# 3) Roboflow SDK로 데이터 다운로드
from roboflow import Roboflow
rf = Roboflow(api_key=os.getenv("ROBOFLOW_API_KEY"))
project = rf.workspace("wiredamagedetect").project("wire_damage_byson")
dataset = project.version(1).download("yolov8")  # v1번 버전 다운로드
print("✅ Dataset YAML:", dataset.data_yaml)

# 4) YOLOv8 모델 가져오기 & 학습
from ultralytics import YOLO
model = YOLO("yolov8n.pt")   # 'n' 또는 's','m' 등 원하시는 크기로 변경 가능

results = model.train(
    data=dataset.data_yaml,   # Roboflow가 생성해 준 data.yaml
    epochs=30,                # epoch 수 조정 가능
    imgsz=640,                # Roboflow 버전에서 설정한 이미지 크기
    batch=8,                  # GPU 메모리에 맞춰 조절
    project="runs/train",     # 결과 저장 폴더
    name="wire_break_exp1"    # 실험 이름
)

# 학습이 끝나면 best.pt 경로와 손실 곡선을 확인해 봅니다
print("✅ Best weights:", results.best)
from IPython.display import Image, display
display(Image(str(results.plot_loss())))
```
